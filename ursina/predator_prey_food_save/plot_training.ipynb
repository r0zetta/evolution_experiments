```python
import json, os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
predator = []
with open("evolution_stats_predator.json", "r") as f:
    predator = json.loads(f.read())

prey = []
with open("evolution_stats_prey.json", "r") as f:
    prey = json.loads(f.read())

if len(predator) < len(prey):
    for n in range(len(prey)-len(predator)):
        predator.append(None)
if len(prey) < len(predator):
    for n in range(len(predator)-len(prey)):
        prey.append(None)
sns.set(style="whitegrid")

data = {"predator":predator, "prey":prey}

fig = plt.figure(figsize=(15,8))
df = pd.DataFrame(data)
df.head()

df = df.rolling(7).mean()
ax = sns.lineplot(data=df, hue="event", style="event", dashes=False, markers=False, linewidth=1.5)
ax.xaxis.set_major_locator(plt.MaxNLocator(5))
plt.show()
```
